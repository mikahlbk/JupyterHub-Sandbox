$u(t,x) = e^{-.2t}\sin(\pi x)$ with $x$ on $[0,1]$, $t\geq 0$.

The first idea for plotting the solution to a PDE is to *show* several discrete snapshots of time and to arrange the plots in an array so we can read from left to right to see the evolution of the solution in time.

```python
import numpy as np
import matplotlib.pyplot as plt
u = lambda t,x: np.exp(-.2*t)*np.sin(np.pi*x)# lambda function to define the PDE
x = np.linspace(0,1,100)# code that gives 100 equally spaced points from 0 to 1
t = np.linspace(0,10,16)# code that gives 16 equally spaced points from 0 to 10
fig, ax = plt.subplots(nrows=4,ncols=4,figsize=(12,8))
counter = 0 # this counter will count through the times
for n in range(4):
    for m in range(4):
        ax[n,m].plot(x,u(t[counter],x), 'b') # plot x vs u(t[counter],x)
        ax[n,m].grid()
        ax[n,m].set_ylim(0,1) # same axis for every plot
        ax[n,m].set_xlabel('x')
        ax[n,m].set_ylabel('u')
        ax[n,m].set_title("time={:.2f}".format(t[counter]))
        counter += 1 # increment the counter
fig.tight_layout()
plt.show()
```

A second idea for plotting the solution to a PDE is to give an interactive plot where we can use a slider to advance (or reverse) time.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive

u = lambda t, x: np.exp(-0.2*t) *  np.sin(np.pi*x)
x = np.linspace(0,1,100)

def plotter(T):
  plt.plot(x , u(T,x), 'b')
  plt.grid()
  plt.ylim(0,1)
  plt.show()

interactive_plot = interactive(plotter, T=(0,20,0.1))
interactive_plot
```
